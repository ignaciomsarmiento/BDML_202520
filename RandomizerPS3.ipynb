```python
import importlib
import subprocess
import sys
import random
import pandas as pd

SEED = 20251125
random.seed(SEED)
```

```python
miembros_equipos = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/03_Presentacion03/_nombres.xlsx")
```

```python
miembros_equipos
```

```python

miembros_equipos["equipo_num"] = (
    miembros_equipos["Equipos Taller 3"]
    .astype(str)
    .str.extract(r"(\d+)", expand=False)
    .dropna()
    .astype(int)
)
```

# Data




## Primer presentador

```python
available_teams = sorted(miembros_equipos['equipo_num'].dropna().unique().astype(int).tolist())
print(f"Available teams: {available_teams}")
```

```python

selected_team = random.choice(available_teams)
print(f"Selected Team: {selected_team}")
```

```python
members_of_selected_team = miembros_equipos[miembros_equipos['equipo_num'] == selected_team]
selected_member = members_of_selected_team.sample(n=1)

print(f"Selected Member: {selected_member['First Name'].iloc[0]} {selected_member['Last Name'].iloc[0]} ({selected_member['Username'].iloc[0]})")
```

```python
# remove selected team
if selected_team in available_teams:
    available_teams.remove(selected_team)
print(f"Updated available teams: {available_teams}")
```
